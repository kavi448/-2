```python
# Устанавливаем порог для удаления столбцов с пропущенными значениями
tbreshold = 0.7 * len(df)  
# Здесь мы вычисляем пороговое значение (tbreshold), которое равно 70% от общего количества строк в DataFrame (len(df)).
# Это означает, что если в столбце будет больше пропущенных значений, чем 30% от общего числа строк, этот столбец будет удален.

# Удаляем столбцы с пропущенными значениями, если их количество превышает установленный порог
df = df.dropna(axis=1, thresh=tbreshold)  
# Метод dropna() удаляет столбцы или строки с пропущенными значениями.
# Сохраняем очищенные данные в новый CSV-файл
df.to_csv('C:\\Users\\Karina\\Desktop\\Интенсив\\clean_train.csv', index=0)  
# Метод to_csv() сохраняет DataFrame в CSV-файл по указанному пути.
# Выводим информацию о очищенном DataFrame
df.info()  
# Метод info() снова выводит информацию о DataFrame после удаления столбцов.
# Подсчитываем количество пропущенных значений в каждом столбце после очистки
print(df.isna().sum())  
# Снова используем метод isna() для проверки наличия пропущенных значений в оставшихся столбцах.
```

```python
# Предполагаем, что df - это ваш датафрейм
df = pd.read_csv('C:\\Users\\Karina\\Desktop\\Интенсив\\clean_train.csv')

# Шаг 1: Заполнение NaN значений единицами
df = df.fillna(1)

# Шаг 2: Кодирование категориальных колонок
le = LabelEncoder()

# Получаем список категориальных колонок
categorical_columns = df.select_dtypes(include=['object']).columns

# Применяем LabelEncoder к каждой категориальной колонке
for col in categorical_columns:
    df[col] = le.fit_transform(df[col].astype(str))

# Выводим первые несколько строк обработанного датафрейма
print(df.head())

# Выводим информацию о датафрейме
print(df.info())

df.to_csv('C:\\Users\\Karina\\Desktop\\Интенсив\\new_clean_train.csv', index = 0)
```

```python
# Загружаем данные из CSV-файла в DataFrame df
df = pd.read_csv('C:\\Users\\Karina\\Desktop\\Интенсив\\new_clean_train.csv')  
# Метод read_csv() используется для чтения данных из файла в формате CSV и загрузки их в DataFrame.
# Указываем полный путь к файлу, который мы хотим загрузить. 

# Настраиваем параметры pandas для использования бесконечностей как NaN
pd.options.mode.use_inf_as_na = True  
# Эта настройка позволяет pandas интерпретировать бесконечные значения (inf, -inf) как пропущенные значения (NaN).
# Это может быть полезно, если ваши данные содержат бесконечности, и вы хотите, чтобы они обрабатывались так же, как и NaN.

# Выводим количество пропущенных значений в каждом столбце DataFrame
print(df.isna().sum())  
# Метод isna() возвращает DataFrame того же размера, что и df, где значения NaN обозначаются как True, а остальные — как False.
```

```python
# Устанавливаем пороговое значение для удаления столбцов
threshold = 0.01 * len(df)  
# Здесь мы вычисляем порог, который равен 1% от общего количества строк в DataFrame df.
# Это значение будет использоваться для определения, сколько нулевых значений (в виде строк '0.0') может содержать столбец,
# прежде чем он будет удален из DataFrame. Если столбец содержит больше нулевых значений, чем этот порог, он будет удален.

# Удаляем столбцы, где количество '0.0' превышает порог
df = df.drop(df.columns[df.apply(lambda col: col.astype(str).eq('0.0').sum() > threshold)], axis=1)  
# Метод apply() применяется к каждому столбцу (col) DataFrame df.
# Внутри apply() мы используем лямбда-функцию, которая:
# 1. Преобразует значения столбца в строки с помощью astype(str).
# 2. Сравнивает каждое значение с '0.0' с помощью eq('0.0'), возвращая Series с True для совпадений и False для остальных значений.
# 3. Суммирует количество True значений (т.е. количество '0.0') с помощью sum().
# Затем мы сравниваем это количество с порогом (threshold).
# Если количество '0.0' в столбце превышает порог, этот столбец будет включен в список для удаления.
# Метод drop() затем удаляет эти столбцы из DataFrame df по указанной оси (axis=1), что означает, что мы работаем со столбцами.

# Сохраняем очищенный DataFrame в новый CSV-файл
df.to_csv('C:\\Users\\Karina\\Desktop\\Интенсив\\new_new_clean_train.csv', index=0)  
```

```python
import pandas as pd  

# Загрузка данных из CSV файла
df = pd.read_csv('C:\\Users\\Karina\\Desktop\\Интенсив\\new_new_clean_train.csv')  
# Путь к файлу был изменен на правильный формат с использованием "/" вместо "\".

# Конвертирование столбца 'Итоговая сумма' в числовой тип данных
df['Итоговая сумма'] = pd.to_numeric(df['Итоговая сумма'], errors='coerce')  

# Вычисление квартили и IQR
q1 = df['Итоговая сумма'].quantile(0.25)  
q3 = df['Итоговая сумма'].quantile(0.75)  
iqr = q3 - q1  

# Вычисление границ для фильтрации выбросов
lower_bound = q1 - 1.5 * iqr  
upper_bound = q3 + 1.5 * iqr  

# Фильтрация данных
df = df[(df['Итоговая сумма'] >= lower_bound) & (df['Итоговая сумма'] <= upper_bound)] 

# Сохранение очищенных данных в новый CSV файл
cleaned_data = 'C:\\Users\\Karina\\Desktop\\Интенсив\\cleaned_data.csv'  
df.to_csv(cleaned_data, index=False)  

print(f'Очищенные данные сохранены в "{cleaned_data}"')  
```
